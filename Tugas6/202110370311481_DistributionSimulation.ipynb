#Poisson Distribution Simulation

##Import required libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

##Set seed for reproducibility

```python
np.random.seed(7)
```

##Simulation parameters

```python
lambda_defect = 3   # Average number of defective items per day
num_days = 30       # Number of simulation iterations (30 days)
```

##Simulate the number of defective items per day using Poisson distribution

```python
defects_per_day = np.random.poisson(lam=lambda_defect, size=num_days)
```

##Display simulation results

```python
print("Simulated number of defective items per day for 30 days:")
print(defects_per_day)
```

##Visualization: Histogram of daily defects

```python
plt.figure(figsize=(10, 5))
sns.histplot(defects_per_day, bins=range(0, max(defects_per_day)+2), kde=False, color='skyblue', edgecolor='black')
plt.title("Histogram of Daily Defective Items (30 Days)")
plt.xlabel("Number of Defects")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

##Visualization: Defect trend over time

```python
plt.figure(figsize=(10, 5))
plt.plot(range(1, num_days+1), defects_per_day, marker='o', linestyle='-', color='red')
plt.title("Trend of Defective Items per Day")
plt.xlabel("Day")
plt.ylabel("Number of Defects")
plt.grid(True)
plt.show()
```

##Statistical analysis

```python
mean_simulation = np.mean(defects_per_day)
print(f"Average from simulation: {mean_simulation:.2f}")
print(f"Theoretical mean (λ): {lambda_defect}")
```
